```python
"""
pandas 基础操作
read_xxx 读取
to_xxx 写入
"""
import pandas as pd

# read_xxx  读取信息
df = pd.read_csv("data/test_list.csv", usecols=['电影名称', '评分', '播放时长', '语言'], encoding="utf-8-sig")

df['金额'] = df['评分'] * 100

# to_csv  保存信息
df.to_csv("data/test_list_new.csv", index=False)

""" 数据的查询 """
# df.head(10) # 显示前10行
# df.tail(10) # 显示最后10行
# df.info() # 显示数据信息
df.describe()  # 显示数据统计信息
```

```python
""" 数据的选择 """
# 选择单列
df['电影名称']

# 多列
df[['电影名称', '评分', '播放时长']]

# 选择行
df.iloc[0:5]  # 基于默认行号选择
df.loc[0:5]  # 基于行索引值选择
```

```python
""" 数据的筛选 """
df[df['评分'] >= 86]

df[(df['电影名称'].isin(['The Shawshank Redemption', 'The Godfather']))]  # 筛选这两条数据

df[df['金额'].between(8600, 9000)]  # 筛选金额在 8600-9000之间的数据 默认包含首尾值(inclusive='left' 可以自定义包含规则)

df[(df['金额'].between(8600, 9000)) & (df['电影名称'].isin(['The Shawshank Redemption', 'The Godfather']))] # 多条件 筛选
```
