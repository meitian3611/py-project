```python
import pandas as pd

pd.set_option("display.max_columns", None)  # 显示所有列
df = pd.read_csv("data/test_list.csv")

### 缺失值
df.isnull().sum()  # 统计缺失值
# df.dropna(inplace=True) # 删除缺失值所在行

### 填充缺失值
df.fillna('---')

### 重复值处理
df.duplicated()  # 检测重复值(默认检测所有列是否全部相等)
df.duplicated(subset=['评分'])  # 检测指定列是否重复
df.drop_duplicates(subset=['评分'], keep='first')  # 删除重复值 first:保留第一个 last:保留最后一个  False:删除所有重复值

### 异常值处理
df[df['评分'] < 86]  # 模拟筛选 不满足条件的异常值
df.drop(df[df['评分'] < 86].index)  # 删除异常值

### 数据格式处理
# df['上映时间'] = df['上映时间'].apply(lambda x: x.replace('-', '/'))
df['上映时间'] = df['上映时间'].str.replace('-', '/')

### 数据排序
df.sort_values(['评分'], ascending=False)  # 是否升序(从小到大) 默认True
df.sort_values(['评分', '上映时间'], ascending=[False, False])  # 支持多列, 评分一样 才去根据上映时间继续排序, 注意先后顺序

### 数据分组
df['总金额'] = 100 * df['评分']

df.groupby(['语言'])['总金额'].count()  # 出现的总个数
df.groupby(['语言'])['总金额'].sum()  # 求和
df.groupby(['语言'])['总金额'].max()  # 求最大
df.groupby(['语言'])['总金额'].min()  # 求最小
df.groupby(['语言'])['总金额'].mean()  # 求平均

df.groupby(['语言'])[['总金额', '评分']].agg(['count', 'sum', 'max', 'min', 'mean'])  # 聚合多个统计操作
df.groupby(['语言']).agg({'总金额': ['count', 'sum', 'max', 'min', 'mean'], '评分': ['max', 'min']}) # 字典格式
```
